```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn
seaborn.set()
```

```python
raw_data = pd.read_csv('1.03. Dummies.csv')
raw_data
```

```python
data = raw_data.copy()
data['Attendance'] = data['Attendance'].map({'Yes': 1, 'No': 0})
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'], y, c=data['Attendance'], cmap='RdYlGn_r')
yhat_no = 0.0014*data['SAT'] + 0.6439
yhat_yes = 0.0014*data['SAT'] + 0.2226 + 0.6439
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='green', label='regression line with < 75% attendance')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='red', label='regression line with > 75% attendance')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```
