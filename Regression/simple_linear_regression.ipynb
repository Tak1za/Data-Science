```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('1.01. Simple linear regression.csv')
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1, y)
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
plt.scatter(x1, y)
# Taking into consideration p-value of constant 
# If p-value is greater than 0.05 it means it is not significantly different from 0
yhat = 0.0017*x1 + 0
fig = plt.plot(x1, yhat, lw=4, c='red', label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```

```python
plt.scatter(x1, y)
# Not taking into consideration p-value of constant.
# If p-value is greater than 0.05 it means it is not significantly different from 0
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1, yhat, lw=4, c='red', label='regression line')
plt.xlabel('SAT', fontsize=20)
plt.ylabel('GPA', fontsize=20)
plt.show()
```
